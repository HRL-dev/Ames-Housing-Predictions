```python
#importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
# Reading in data
ames_houses = pd.read_csv('../datasets/train.csv')

ames_house = pd.read_csv('../datasets/test.csv')
```

```python
# Removing extreme outliers from house area columns

ames_houses = ames_houses[ames_houses['Gr Liv Area'] < 5000]

ames_house = ames_house[ames_house['Gr Liv Area'] < 5000]
```

```python
features = ['Garage Cars',
            'Garage Qual',
            'Garage Cond',
            'Garage Type',
            'Garage Finish',
           'Garage Area']

ames_train = ames_houses[features].copy()

ames_test = ames_house[features].copy()
```

```python
feat = ['Garage Cars',
            'Garage Qual',
            'Garage Cond',
            'Garage Type',
            'Garage Finish',
           'Garage Area',
            'SalePrice']

ames_feat = ames_houses[feat].copy()
```

```python
ames_train.isnull().sum()
```

```python
ames_test.isnull().sum()
```

```python
# We are going to fill the null values with 0, because we need the X_train to maintain its shape, but
# we can't have null values when we run our model.

ames_train = ames_train.fillna(0)
```

```python
ames_train.isnull().sum()
```

```python
ames_test = ames_test.fillna(0)
```

```python
ames_train.dtypes
```

```python
# We are dummifying the ordinal values in order to use them in our model.

X_train = pd.get_dummies(data=ames_train, columns=['Garage Qual',
                                                  'Garage Cond', 'Garage Type',
                                                  'Garage Finish'], drop_first=True)
X_test = pd.get_dummies(data=ames_test, columns=['Garage Qual',
                                                  'Garage Cond', 'Garage Type',
                                                  'Garage Finish'], drop_first=True)
ames_feat = pd.get_dummies(data=ames_feat, columns=['Garage Qual',
                                                  'Garage Cond', 'Garage Type',
                                                  'Garage Finish'], drop_first=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Because X_train and X_test are different shapes and we need them to be the same, we are looking for the
#columns that are in one that are not in the other.

for column in X_train.columns:
    if column not in X_test.columns:
        print(column)
```

```python
for column in X_test.columns:
    if column not in X_train.columns:
        print(column)
```

```python
X_train['Garage Qual_Ex'].value_counts()
```

Dropping the column that is in X_train and not X_test seems best for matching the shapes.

```python
X_train = X_train.drop(columns=['Garage Qual_Ex'])
```

```python
X_train.shape
```

```python
# Defining the target variable:

y = ames_houses['SalePrice']
```

```python
y.shape
```

```python
# Making a correlation heatmap to determine which features are most important for predicting price.
plt.figure(figsize = (5, 5))

sns.heatmap(ames_feat.corr()[['SalePrice']],
            vmin = -1,
            vmax = 1,
            annot = True,
            cmap = 'mako');

plt.title('The Correlation between Sale Price and Garage Features', fontsize = 15)

plt.savefig('garage_corr.png')
```

```python
# Examining the strong relationship between the increasing area of the garage and the resulting increasing
# sale price.

sns.lmplot(x='Garage Area', y='SalePrice', data=ames_feat, palette = 'mako', ci=None)

plt.title('The Effect of Garage Area on Sale Price', fontsize = 15);

plt.savefig('area_price.png')
```

```python
ames_feat['Unfinished Garage'] = X_train['Garage Finish_Unf']
```

```python
plt.bar(ames_feat['Unfinished Garage'], ames_feat['SalePrice'])

plt.title('The Effect of an Unfinished Garage on Sale Price', fontsize = 15)

plt.xticks(ticks=[0, 1], labels = ['Finished Garage', 'Unfinished Garage'])

plt.xlabel('Garage Finish')

plt.ylabel('Sale Price');

plt.savefig('unfinished_price.png')
```

```python
# Making a new column based on strong correlation:

X_train['garage_space'] = X_train['Garage Cars'] * X_train['Garage Area']

X_test['garage_space'] = X_test['Garage Cars'] * X_test['Garage Area']

ames_feat['Garage Space'] = ames_feat['Garage Cars'] * ames_feat['Garage Area']
```

```python
# Plotting the new column vs. Sale Price to see the effect:

sns.lmplot(x='Garage Space', y='SalePrice', data=ames_feat, ci=None)

plt.title('The Effect of Total Garage Space on Sale Price', fontsize = 15);

plt.savefig('space_price.png')
```

```python
# Instantiating model and finding the mean of the cross value score:
lr = LinearRegression()

cross_val_score(lr, X_train, y).mean()
```

```python
# Train-test splitting the data:

X_train_1, X_test_1, y_train, y_test = train_test_split(X_train, y, random_state = 142)
```

```python
# Fitting the linear regression model:

lr.fit(X_train_1, y_train)
```

```python
lr.score(X_train_1, y_train)
```

```python
lr.score(X_test_1, y_test)
```

```python
y_preds = lr.predict(X_train)
```

```python
RMSE = np.sqrt(metrics.mean_squared_error(y, y_preds))

RMSERMSE = np.sqrt(metrics.mean_squared_error(y, y_preds))

RMSE
```

So, the scores for our model were not super impressive, and our root mean square error is too high. How could we change our model to make it better while staying true to our client's interests? The first step might be to add some columns from the original dataframe that are very strongly correlated with the sale price column.

```python
plt.figure(figsize=(5, 12))

sns.heatmap(ames_houses._get_numeric_data().corr()[['SalePrice']],
            vmin = -1,
            vmax = 1,
            annot = True,
            cmap = 'mako')

plt.title('Correlations: Sale Price vs All Other Numeric Columns');

plt.savefig('numeric_corrs.png')
```

```python
# Adding in the columns with the strongest correlation:

X_train['Overall Qual'] = ames_houses['Overall Qual']

X_test['Overall Qual'] = ames_house['Overall Qual']
```

```python
X_train['Gr Liv Area'] = ames_houses['Gr Liv Area']

X_test['Gr Liv Area'] = ames_house['Gr Liv Area']
```

```python
# We'll also add an interaction column between those two columns:

X_train['overall_living'] = X_train['Gr Liv Area'] * X_train['Overall Qual']

X_test['overall_living'] = X_test['Gr Liv Area'] * X_test['Overall Qual']
```

```python
# More columns:

X_train['Total Bsmt SF'] = ames_houses['Total Bsmt SF']

X_test['Total Bsmt SF'] = ames_house['Total Bsmt SF']
```

```python
# Filling new nulls with 0:

X_train['Total Bsmt SF'] = X_train['Total Bsmt SF'].fillna(0)

X_test['Total Bsmt SF'] = X_test['Total Bsmt SF'].fillna(0)
```

```python
X_train['1st Flr SF'] = ames_houses['1st Flr SF']

X_test['1st Flr SF'] = ames_house['1st Flr SF']
```

```python
# and an interaction column:

X_train['low_sq_ftg'] = X_train['1st Flr SF'] * X_train['Total Bsmt SF']

X_test['low_sq_ftg'] = X_test['1st Flr SF'] * X_test['Total Bsmt SF']
```

```python
# more columns:

X_train['Year Built'] = ames_houses['Year Built']

X_test['Year Built'] = ames_house['Year Built']
```

```python
X_train['Year Remod/Add'] = ames_houses['Year Remod/Add']

X_test['Year Remod/Add'] = ames_house['Year Remod/Add']
```

```python
# and an interaction column:

X_train['years'] = X_train['Year Built'] * X_train['Year Remod/Add']

X_test['years'] = X_test['Year Built'] * X_test['Year Remod/Add']
```

```python
# Now, we'll re-split the data, re-fit the model, and try again:

X_train_1, X_test_1, y_train, y_test = train_test_split(X_train, y, random_state = 142)
```

```python
lr.fit(X_train_1, y_train)
```

```python
cross_val_score(lr, X_train, y).mean()
```

```python
lr.score(X_train_1, y_train)
```

```python
lr.score(X_test_1, y_test)
```

```python
y_preds = lr.predict(X_train)
```

```python
RMSE = np.sqrt(metrics.mean_squared_error(y, y_preds))

RMSE
```

So, we've improved our model by twenty points, and reduced our root mean square error by 15,000 - meaning our predictions are significantly close.
